```python
def open_dataset(filename, sep=';'):
    data = []
    f = open(filename, 'r')
    win = []
    for i in f.readlines():
        i = i.strip()
        if i != '--':
            win.append([float(x) for x in i.split(',')])
        else:
            data.append(win)
            win = []
    return data
```

```python
dataset = open_dataset('log_file.txt')
```

```python
print len(dataset)
```

```python
from collections import defaultdict
win = 5

fig = figure(figsize=(20,10))

ax = fig.add_subplot(1,4, 1)

s_lens = []
for p in range(20):
    ax = fig.add_subplot(5,4, p+1)
    
    clusters = defaultdict(list)
    
    for x, y in zip([int(j[0]) for j in dataset[p]], [float(j[1]) for j in dataset[p]]):
        clusters[x].append(y)
    
    lens = []
    for k, v in clusters.iteritems():
        
        var = np.var(v)
        
        lens.append(len(v))

        ax.scatter(k, np.var(v))
        ax.set_xlim((-1,10))
        ax.set_xticks(range(-1,11,2))
        ax.set_xlabel('Cluster ID')
        ax.set_ylabel('Smallest Eigen proj')
    
    s_lens.append(np.mean(lens))
    print "In plot {} mean length {}".format(p, np.mean(lens))
plt.tight_layout()
plt.show()

fig = figure(figsize=(12,3))
plt.hist(s_lens)
plt.show()
```

#### PLOT ALL VARS

```python
all_vars = []
lens = []

count = 0
nclusters = 0
for p in range(len(dataset)):
    clusters = defaultdict(list)
    for x, y in zip([int(j[0]) for j in dataset[p]], [float(j[1]) for j in dataset[p]]):
        clusters[x].append(y)
    
    for k, v in clusters.iteritems():
        var = np.var(v)
        all_vars.append(var)
        lens.append(len(v))
        nclusters +=1
    
    count +=1

print "Total number of windows: {}".format(count)
print "Total number of clusters: {}".format(nclusters)
plt.plot(all_vars)
plt.show()
```

```python
dataset = open_dataset('log_file_jaccard.txt')
```

```python
from collections import defaultdict

data = dataset[3]

points = defaultdict(list)
for x,y in data:
    points[x].append(y)

for k in points.keys():
    plt.plot(range(len(points[k])), points[k], label="Cluster #{}".format(int(k)))
plt.ylabel('angle')
plt.xlabel('clusters')
plt.legend(loc='best')
plt.show()
```
